```python
import numpy as np
import matplotlib.pyplot as plt

import sys
sys.path.append('../')
from scint_tools import scint_utils

plt.rc('font', family='serif')
plt.rcParams.update({'font.size': 16})
```

```python
fig, ax = plt.subplots(1,1, figsize=(16,20))
i = 0
for h5_file in open("fnames.txt"):
    mydata = scint_utils.scintdata(h5_file.strip())
    mydata.RFIsub()
    print (i)
    if i==0:
        saveobsdate = mydata.obs_date
    satdlink_proc_pow_c1 = mydata.satdlink_proc_pow_c1-(i*20) #To stack the plots !
    ax.plot(mydata.hrs_full, satdlink_proc_pow_c1)
    ax.annotate(str(mydata.obs_date), xy = (34, satdlink_proc_pow_c1[-1]+1))
    i = i+1

first_hr = int(mydata.hrs_full[0])+1
last_hr = int(mydata.hrs_full[-1])+1
xticks = np.arange(first_hr,last_hr, 2)
xlabels = [(str(int(xtick/24))+"d"+str(xtick%24).zfill(2)+"h") for xtick in xticks]
ax.set_xticks(xticks, labels=xlabels)
ax.set_yticks([])
ax.grid()
ax.set_xlim()
ax.set_title("RFI subtracted scintillation time series, Freq = {:.2f} MHz".format(mydata.freq_1))

ax.set_xlabel("Local time (IST)")
fig.savefig("JoyDivision_"+saveobsdate+".pdf", dpi=300, bbox_inches='tight')
plt.close()
```
